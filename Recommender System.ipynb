```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
movie = pd.read_csv('data/Disney.csv')
```

```python
movie.columns
```

```python
moviesIMP = movie[['show_id','title','director','cast','listed_in','description']]
```

```python
moviesIMP
```

```python
moviesIMP.dropna(inplace=True)
moviesIMP
```

```python
moviesIMP['tag'] = moviesIMP['director'] + ' ' + moviesIMP['cast'] + ' ' + moviesIMP['listed_in'] + ' ' + moviesIMP['description']
moviesIMP.head()
```

```python
moviesIMP['title2'] = moviesIMP['title']
moviesIMP.head()
```

```python
def remove_spaces(word):
    if isinstance(word, str):
        return word.replace(" ", "")
    return word

moviesIMP['listed_in'] = moviesIMP['listed_in'].apply(remove_spaces)
moviesIMP['director'] = moviesIMP['director'].apply(remove_spaces)
moviesIMP['cast'] = moviesIMP['cast'].apply(remove_spaces)
moviesIMP['title2'] = moviesIMP['title2'].apply(remove_spaces)


moviesIMP.head()
```

```python
moviesIMP['tag'] = moviesIMP['director'] + ' ' + moviesIMP['cast'] + ' ' + moviesIMP['listed_in'] + ' ' + moviesIMP['description'] + ' ' + moviesIMP['title2']
moviesIMP.head()
```

```python
FinalMovies = moviesIMP[['show_id', 'title', 'tag']]
FinalMovies.head()
```

```python
import nltk
from nltk.stem import PorterStemmer

ps=PorterStemmer()
```

```python
def stems(text):
    l = []
    for i in text.split():
        l.append(ps.stem(i))
    return " ".join(l)
```

```python
FinalMovies['tag'] = FinalMovies['tag'].apply(stems)
FinalMovies.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000)
```

```python
vectors = cv.fit_transform(FinalMovies['tag']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)

similarity
```

```python
def recommend(movie):
    if movie not in FinalMovies['title'].values:
        print(f"Movie '{movie}' not found in the dataset.")
        return
    movie_index = FinalMovies[FinalMovies['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])
    for i in movies_list[1:6]:
        print(FinalMovies.iloc[i[0]].title)
```

```python
recommend('Coco')
```
